```python
# First of all, salary is a sensitive infomation and some people do not want to provide it. 
# So there is missing value in column salary.
# By looking result of cell [5] that we know our raw dataframe has 51392 records and 154 fields. 
# And you could see the result of this cell is stats from column.
# Salary which actual have only 12891 real value. 
# Although we can replace missing value by mean value or mode value 
# but I want a precisely analysis result and avoid some confused situation like
# 2 rows have same salary value but other fields are totally different.
#Then I'll remove rows which do not contain Salary real value.

raw_df['Salary'].describe()
```

```python
# Remove rows have NaN Salary value

real_salary_df = raw_df.dropna(subset=['Salary'])
real_salary_df.head()
```

```python
def create_dummy_df(df, cat_cols, dummy_na):
    '''
    INPUT:
    df - pandas dataframe with categorical variables you want to dummy
    cat_cols - list of strings that are associated with names of the categorical columns
    dummy_na - Bool holding whether you want to dummy NA vals of categorical columns or not
    
    OUTPUT:
    df - a new dataframe that has the following characteristics:
            1. contains all columns that were not specified as categorical
            2. removes all the original columns in cat_cols
            3. dummy columns for each of the categorical columns in cat_cols
            4. if dummy_na is True - it also contains dummy columns for the NaN values
            5. Use a prefix of the column name with an underscore (_) for separating 
    '''
    for col in  cat_cols:
        try:
            # for each cat add dummy var, drop original column
            df = pd.concat([df.drop(col, axis=1), pd.get_dummies(df[col], prefix=col, prefix_sep='_', drop_first=True, dummy_na=dummy_na)], axis=1)
        except:
            continue
    return df
```
